# Is comedy dead? Rotten Tomatoes database EDA 

## Introduction

The main goal of this study is to determine whether people still enjoy comedy movies.

Questions to explore:

1. How have audience ratings for comedy movies changed over the years?
2. Are comedy movies rated higher or lower compared to other genres?
3. Do critics and audiences rate comedy movies differently?
4. Are comedy movies still widely watched?
5. Which studios produce the highest-rated comedy movies?
6. Do longer comedy movies receive higher or lower ratings?

### Importing libraries and data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'Rotten Tomatoes Movies.csv')
```

### Data cleaning

```python
df['rating'] = df['rating'].str.replace('R)', 'R')
df['rating'] = df['rating'].str.replace('PG-13)', 'PG-13')
```

### Basic data exploration

```python
df.sample(5)
```

```python
df.sort_values('audience_count', ascending=True)
```

```python
df.info()
```

```python
df['in_theaters_date'] = pd.to_datetime(df['in_theaters_date'])
df['on_streaming_date   '] = pd.to_datetime(df['on_streaming_date'])
```

```python
df.nunique()
```

### Descriptive statistics for numerical data

```python
popular_movies = df[df['audience_count'] >= 100000].copy()
```

```python
popular_movies.describe(include=[np.number]).T
```

```python
popular_movies.corr(numeric_only=True)
```

```python
sns.pairplot(popular_movies)
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(popular_movies['tomatometer_rating'], bins=20, color='orange', label='Critics Rating', alpha=1)
sns.histplot(popular_movies['audience_rating'], bins=20, color='navy', label='Audience Rating', alpha=0.5)

plt.title('Comparison of Critics and Audience Ratings', fontsize=14)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.legend()

plt.show()
```

```python
popular_movies.groupby('rating')[['tomatometer_rating', 'audience_rating']].mean()
```
